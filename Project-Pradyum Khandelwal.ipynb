```python
# Import necessary libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm


# Define the independent variables (features) and the dependent variable (target)
X = cd1[['mileage', 'vehicle_age', 'engine']]
y = cd1['selling_price']
X.fillna(X.mean(), inplace=True)
y.fillna(y.mean(), inplace=True)


# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train the model using sklearn's LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)

# Print the coefficients
print('Intercept:', model.intercept_)
print('Coefficients:', model.coef_)

# Make predictions
y_pred = model.predict(X_test)

# Evaluate the model
from sklearn.metrics import mean_squared_error, r2_score
print('Mean Squared Error:', mean_squared_error(y_test, y_pred))
print('R-squared:', r2_score(y_test, y_pred))

# Now let's do the same using statsmodels to get more detailed statistics
X_train_sm = sm.add_constant(X_train)
model_sm = sm.OLS(y_train, X_train_sm).fit()
print(model_sm.summary())

# Make predictions on test set
X_test_sm = sm.add_constant(X_test)
y_pred_sm = model_sm.predict(X_test_sm)

# Evaluate the model
print('Mean Squared Error (Statsmodels):', mean_squared_error(y_test, y_pred_sm))
print('R-squared (Statsmodels):', r2_score(y_test, y_pred_sm))
```
